```python
import numpy as np
from scipy.stats import norm

def runs_test_AB(sequence, alpha=0.05):
    seq = np.array(list(sequence))
    binary_seq = np.where(seq == 'A', 1, 0)
    R = 1
    for i in range(1, len(binary_seq)):
        if binary_seq[i] != binary_seq[i-1]:
            R += 1
    n1 = np.sum(binary_seq == 1)
    n2 = np.sum(binary_seq == 0)
    mu_R = (2 * n1 * n2) / (n1 + n2) + 1
    sigma_R = np.sqrt((2 * n1 * n2 * (2 * n1 * n2 - n1 - n2)) / ((n1 + n2)**2 * (n1 + n2 - 1)))
    z = norm.ppf(1 - alpha/2)
    g1 = int(np.floor(mu_R - z * sigma_R - 0.5))
    g2 = int(np.floor(mu_R + z * sigma_R + 0.5))
    if R <= g1 or R >= g2:
        return "Reject"
    else:
        return "Cannot reject"

sequence = "AAABBAAAABAABAAAABBBB"
print(runs_test_AB(sequence))
```
